Les variables du dataset correspondent à des caractéristiques morphologiques
mesurées sur les noyaux de cellules issues d’images de biopsies mammaires.
Ces mesures décrivent la taille, la forme et la complexité du contour des cellules.

Traduction pédagogique des variables

| Colonne            | Traduction                 | Définition simple                                               |
| ------------------ | -------------------------- | --------------------------------------------------------------- |
| Diagnosis          | Diagnostic                 | Indique si la tumeur est bénigne (B) ou maligne (M).            |
| radius1            | Rayon moyen                | Taille moyenne du noyau de la cellule observée.                 |
| texture1           | Texture moyenne            | Variation d’intensité dans l’image (aspect granuleux du tissu). |
| perimeter1         | Périmètre moyen            | Longueur du contour du noyau cellulaire.                        |
| area1              | Aire moyenne               | Surface du noyau cellulaire.                                    |
| smoothness1        | Lissage moyen              | Régularité du contour du noyau.                                 |
| compactness1       | Compacité moyenne          | Mesure de la compacité : relation entre périmètre et surface.   |
| concavity1         | Concavité moyenne          | Profondeur moyenne des creux dans le contour du noyau.          |
| concave_points1    | Points concaves moyens     | Nombre de zones creusées dans le contour.                       |
| symmetry1          | Symétrie moyenne           | Degré de symétrie de la cellule.                                |
| fractal_dimension1 | Dimension fractale moyenne | Complexité géométrique du contour du noyau.                     |


Variables de variabilité (erreur standard)

| Colonne            | Traduction                        | Définition                                    |
| ------------------ | --------------------------------- | --------------------------------------------- |
| radius2            | Variabilité du rayon              | Variation du rayon entre différentes mesures. |
| texture2           | Variabilité de texture            | Variation de la texture observée.             |
| perimeter2         | Variabilité du périmètre          | Variation du contour du noyau.                |
| area2              | Variabilité de surface            | Variation de la taille du noyau.              |
| smoothness2        | Variabilité du lissage            | Variation de la régularité du contour.        |
| compactness2       | Variabilité de compacité          | Variation de la compacité du noyau.           |
| concavity2         | Variabilité de concavité          | Variation des creux du contour.               |
| concave_points2    | Variabilité des points concaves   | Variation du nombre de creux.                 |
| symmetry2          | Variabilité de symétrie           | Variation de la symétrie de la cellule.       |
| fractal_dimension2 | Variabilité de dimension fractale | Variation de la complexité du contour.        |


Variables "pire cas" (valeur maximale)
| Colonne            | Traduction                  | Définition                          |
| ------------------ | --------------------------- | ----------------------------------- |
| radius3            | Rayon maximal               | Plus grand rayon mesuré.            |
| texture3           | Texture maximale            | Texture la plus marquée observée.   |
| perimeter3         | Périmètre maximal           | Plus grand contour observé.         |
| area3              | Surface maximale            | Plus grande surface du noyau.       |
| smoothness3        | Lissage maximal             | Régularité maximale observée.       |
| compactness3       | Compacité maximale          | Compacité la plus élevée observée.  |
| concavity3         | Concavité maximale          | Creux les plus profonds du contour. |
| concave_points3    | Points concaves maximum     | Nombre maximum de creux observés.   |
| symmetry3          | Symétrie maximale           | Symétrie la plus élevée observée.   |
| fractal_dimension3 | Dimension fractale maximale | Complexité maximale du contour.     |
