```python
#Atelier pratique Groupe 1se

#PARTIE 1 EDA
# 📚 Importation des bibliothèques/modules
import pandas as pd   # bibliothèque utilisée pour manipuler des tableaux de données
import os             # bibliothèque permettant de manipuler les chemins de fichiers
import seaborn as sns # seaborn est une bibliothèque Python utilisée pour créer des graphiques statistiques plus lisibles que matplotlib
import matplotlib.pyplot as plt   # matplotlib permet d'afficher les graphiques
import numpy as np                  # numpy permet de faire des calculs statistiques 
import os 

from ucimlrepo import fetch_ucirepo
```

```python
# 📥 Chargement du dataset Breast Cancer Wisconsin (Diagnostic)

breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)
```

```python
# 📊 Récupérer les variables explicatives
X = breast_cancer_wisconsin_diagnostic.data.features

# 🎯 Récupérer la variable cible
y = breast_cancer_wisconsin_diagnostic.data.targets

# 🔗 Fusionner les deux
df = pd.concat([X, y], axis=1)

# ⚠️ Le package ucimlrepo renomme les colonnes avec les suffixes 1, 2 et 3
# au lieu des noms originaux du dataset UCI : mean, se (standard error) et worst.
# Nous allons donc restaurer les noms scientifiques.

df.columns = (
    df.columns
    # remplacer les colonnes qui se terminent par "1" par "_mean"
    .str.replace(r'1$', '_mean', regex=True) # $ designe la fin du mot

    # remplacer les colonnes qui se terminent par "2" par "_se"
    .str.replace(r'2$', '_se', regex=True)

    # remplacer les colonnes qui se terminent par "3" par "_worst"
    .str.replace(r'3$', '_worst', regex=True)
)

#Etape d'Exportation à conserver svp
#df.to_excel(r"C:\Users\audec\Downloads\breast_cancer_dataset.xlsx", index=False)
# ➜ enlever le # pour exporter le dataframe en fichier Excel dans le dossier Downloads pour faire ensuite les transformations

# Vérifier si le fichier existe bien dans Downloads
#os.path.exists(r"C:\Users\audec\Downloads\breast_cancer_dataset.xlsx")
```

Les variables du dataset correspondent à des caractéristiques morphologiques
mesurées sur les noyaux de cellules issues d’images de biopsies mammaires.
Ces mesures décrivent la taille, la forme et la complexité du contour des cellules.

| Suffixe | Signification | Interprétation |
|---|---|---|
| mean  | moyenne        | valeur moyenne de la mesure                                                                                      |
| se    | standard error | erreur standard de la moyenne = SE mesure l’incertitude de la moyenne, peut être représenté comme barre d’erreur |
| worst | worst value    | moyenne des 3 plus grandes valeurs                                                                               |

Traduction pédagogique des variables

| Colonne               | Traduction                 | Définition simple                                               |
| ----------------------| -------------------------- | --------------------------------------------------------------- |
| Diagnosis             | Diagnostic                 | Indique si la tumeur est bénigne (B) ou maligne (M).            |
| radius mean           | Rayon moyen                | Taille moyenne du noyau de la cellule observée.                 |
| texture mean          | Texture moyenne            | Variation d’intensité dans l’image (aspect granuleux du tissu). |
| perimeter mean        | Périmètre moyen            | Longueur du contour du noyau cellulaire.                        |
| area mean             | Aire moyenne               | Surface du noyau cellulaire.                                    |
| smoothness mean       | Lissage moyen              | Régularité du contour du noyau.                                 |
| compactness mean      | Compacité moyenne          | Mesure de la compacité : relation entre périmètre et surface.   |
| concavity mean        | Concavité moyenne          | Profondeur moyenne des creux dans le contour du noyau.          |
| concave_points mean   | Points concaves moyens     | Nombre de zones creusées dans le contour.                       |
| symmetry mean         | Symétrie moyenne           | Degré de symétrie de la cellule.                                |
| fractal_dimension mean| Dimension fractale moyenne | Complexité géométrique du contour du noyau.                     |

```python
#Histogramme d’une variable (analyse univariée)
plt.figure(figsize=(8,5))

sns.histplot(df["radius_mean"], bins=30, kde=True)

plt.title("Distribution du rayon moyen du noyau")
plt.xlabel("radius_mean")
plt.ylabel("Nombre d'observations")

plt.show()
```

Cette histogramme permet de voir :

distribution

asymétrie

valeurs extrêmes

```python
# Boxplot : Comparaison de benign vs malignant
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Diagnosis",
    y="radius_mean",
    data=df
)

plt.title("Rayon moyen selon le diagnostic")

plt.show()
```

Observation du boxplot : Malignant → valeurs plus grandes

```python
# Visualisation de l’erreur standard (SE)

plt.figure(figsize=(8,5))

sns.pointplot(
    x="Diagnosis",
    y="radius_mean",
    data=df,
    errorbar="se"
)

plt.title("Rayon moyen avec erreur standard")

plt.show()
```

incertitude autour de la moyenne

```python
# Comparaison de mean et worst
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="radius_mean",
    y="radius_worst",
    hue="Diagnosis",
    data=df
)

plt.title("Comparaison rayon moyen vs pire rayon")

plt.show()
```

```python
# Corrélation : Heatmap de corrélation (très importante)

plt.figure(figsize=(12,10))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    cmap="bwr",
    center=0
)

plt.title("Matrice de corrélation")

plt.show()
```

```python
#On reprend le calcul de la matrice de corrélation precedente : corr = df.corr(numeric_only=True)

#Corrélations > 0.5
corr_pairs = corr.unstack().sort_values(ascending=False)
corr_sup_05 = corr_pairs[(corr_pairs > 0.5) & (corr_pairs < 1)]
corr_sup_05.head(20)
```

```python
#On reprend le calcul de la matrice de corrélation precedente : corr = df.corr(numeric_only=True)

#Corrélations < -0.5
corr_neg = corr_pairs[corr_pairs < -0.5]
corr_neg.head(20)
```

Aucune corrélation négative forte dans le dataframe.

Ce heatmap permet de voir :

variables redondantes

variables importantes
