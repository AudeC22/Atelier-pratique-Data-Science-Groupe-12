```python
#Histogramme d’une variable (analyse univariée)
plt.figure(figsize=(8,5))

sns.histplot(df["radius_mean"], bins=30, kde=True)

plt.title("Distribution du rayon moyen du noyau")
plt.xlabel("radius_mean")
plt.ylabel("Nombre d'observations")

plt.show()
```

```python
# 🔧 Fonction qui trace la distribution d'une variable
def plot_distribution(variable):

    data = df[variable].dropna()

    # statistiques
    mean = data.mean()
    std = data.std()
    median = data.median()

    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)

    plt.figure(figsize=(14,6))

    # histogramme
    sns.histplot(
        data,
        bins=30,
        stat="density",
        alpha=0.3,
        label="Histogramme"
    )

    # KDE réelle
    sns.kdeplot(
        data,
        linewidth=3,
        label="Distribution réelle (KDE)"
    )

    # 🔶 courbe gaussienne théorique
    if std > 0:
        x = np.linspace(data.min(), data.max(), 200)
        plt.plot(
            x,
            norm.pdf(x, mean, std),
            color="orange",
            linewidth=3,
            label="Gaussienne théorique"
        )

    # moyenne
    plt.axvline(mean, linestyle="--", label="Moyenne")

    # médiane
    plt.axvline(median, linestyle="-", label="Médiane")

    # quartiles
    plt.axvline(q1, linestyle=":", label="Q1")
    plt.axvline(q3, linestyle=":", label="Q3")

    # écart-type
    plt.axvline(mean - std, linestyle="--", label="μ - σ")
    plt.axvline(mean + std, linestyle="--", label="μ + σ")

    # titre
    plt.title(f"Distribution de {variable}", fontsize=16)

    plt.xlabel(variable)
    plt.ylabel("Densité")

    plt.legend()
    plt.grid(alpha=0.3)

    plt.show()
```

```python
# Étape 1 — toutes les variables mean

mean_cols = [c for c in df.columns if c.endswith("_mean")]

for col in mean_cols:
    plot_distribution(col)
```

```python
# Étape 2 — toutes les variables se
se_cols = [c for c in df.columns if c.endswith("_se")]

for col in se_cols:
    plot_distribution(col)
```

```python
# Étape 3 — toutes les variables worst

worst_cols = [c for c in df.columns if c.endswith("_worst")]

for col in worst_cols:
    plot_distribution(col)
```

Observation du boxplot : Malignant → valeurs plus grandes

```python
# on va faire 3 boxplots séparés :

#_ toutes les colonnes _mean ensemble

#_ toutes les colonnes _se ensemble

#_ toutes les colonnes _worst ensemble

# 🧼 Sélection des colonnes par suffixe  # ✅ On récupère les colonnes mean / se / worst automatiquement
mean_cols = [c for c in df.columns if c.endswith("_mean")]  # ✅ Liste des colonnes qui finissent par "_mean"
se_cols = [c for c in df.columns if c.endswith("_se")]  # ✅ Liste des colonnes qui finissent par "_se"
worst_cols = [c for c in df.columns if c.endswith("_worst")]  # ✅ Liste des colonnes qui finissent par "_worst"

# 🔎 Vérification rapide  # ✅ On vérifie que les listes ne sont pas vides (sinon le graph sera impossible)
print("Nombre de colonnes _mean :", len(mean_cols))  # ✅ Affiche combien de colonnes mean on a
print("Nombre de colonnes _se :", len(se_cols))  # ✅ Affiche combien de colonnes se on a
print("Nombre de colonnes _worst :", len(worst_cols))  # ✅ Affiche combien de colonnes worst on a

# 🧱 Fonction pour tracer un boxplot propre  # ✅ On crée une fonction réutilisable pour éviter de répéter 3 fois le même code
def plot_group_boxplots(cols, title):  # ✅ cols = liste de colonnes à tracer ; title = titre du graphique
    data = df[cols]  # ✅ On extrait uniquement les colonnes voulues du dataframe df
    plt.figure(figsize=(18, 5))  # ✅ On crée une figure large pour que les noms de variables restent lisibles
    plt.boxplot(  # ✅ On crée le boxplot multi-colonnes
        [data[c].dropna().values for c in cols],  # ✅ On convertit chaque colonne en liste de valeurs (sans NaN)
        labels=cols,  # ✅ On met les noms des colonnes en labels sur l’axe X
        showfliers=True  # ✅ On affiche les outliers (points extrêmes) comme dans ton exemple
    )
    plt.title(title)  # ✅ On ajoute le titre du graphique
    plt.ylabel("Valeurs")  # ✅ On ajoute le label de l’axe Y
    plt.xticks(rotation=45, ha="right")  # ✅ On incline les noms de colonnes pour éviter qu’ils se chevauchent
    plt.grid(True, axis="y", linestyle="--", alpha=0.4)  # ✅ On ajoute une grille légère pour lire les quartiles
    plt.tight_layout()  # ✅ On ajuste automatiquement pour éviter que les labels soient coupés
    plt.show()  # ✅ On affiche le graphique

# 📦 Graphique 1 : toutes les variables "mean"  # ✅ On trace toutes les moyennes ensemble
plot_group_boxplots(mean_cols, "Boxplots — Variables _mean (moyennes)")  # ✅ Appel de la fonction pour le groupe mean

# 📦 Graphique 2 : toutes les variables "se"  # ✅ On trace toutes les erreurs standard ensemble
plot_group_boxplots(se_cols, "Boxplots — Variables _se (standard error)")  # ✅ Appel de la fonction pour le groupe se

# 📦 Graphique 3 : toutes les variables "worst"  # ✅ On trace toutes les valeurs worst ensemble
plot_group_boxplots(worst_cols, "Boxplots — Variables _worst (pire valeur)")  # ✅ Appel de la fonction pour le groupe worst
```
