```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Paramètres de la loi binomiale
n, p = 10, 0.3

# Valeurs possibles de X
x = np.arange(0, n+1)

# Probabilités P(X = k)
pmf = binom.pmf(x, n, p)

# Affichage
plt.bar(x, pmf)
plt.title("Loi binomiale")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.show()
```

On ne voit pas bien une tendance vers loi normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Paramètres de la loi binomiale
n, p = 100, 0.3

# Valeurs possibles de X
x = np.arange(0, n+1)

# Probabilités P(X = k)
pmf = binom.pmf(x, n, p)

# Affichage
plt.bar(x, pmf)
plt.title("Loi binomiale")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.show()
```

On voir bien la tendance vers une loi normale

```python
# TCL = Theorem Central Limit
samples = [np.mean(np.random.uniform(0, 1, 30)) for _ in range(1000)]
plt.hist(samples, bins=30)
plt.title("Distribution des moyennes (TCL)")
```
