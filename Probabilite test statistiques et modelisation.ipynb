```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Paramètres de la loi binomiale
n, p = 20, 0.3

# Valeurs possibles de X
x = np.arange(0, n+1)

# Probabilités P(X = k)
pmf = binom.pmf(x, n, p)

# Affichage
plt.bar(x, pmf)
plt.title("Loi binomiale")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.show()
```
